# Exercício 1

Carregue o dataset Iris anexo nesta aula;

Remova a coluna Id;

Use a função countplot() para mostrar a frequência das três espécies de Iris do dataset.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Iris.csv')
```

```python
df.head()
```

```python
df.drop(columns = 'Id', inplace = True)
```

```python
df
```

```python
sns.countplot(x = 'Species', data = df)
plt.title('Frequência de cada uma das espécies')
plt.ylabel('Frequency')
plt.show()
```

# Exercício 2

Faça um scatter plot para mostrar a relação entre 'SepalLengthCm' e 'SepalWidthCm';

Faça um scatter plot para mostrar a relação entre 'PetalLengthCm' e 'PetalWidthCm'.

```python
sns.scatterplot(
    x = 'SepalLengthCm',
    y = 'SepalWidthCm',
    data = df,
    hue = 'Species',
    alpha = 0.5
    );
```

```python
sns.scatterplot(
    x = 'PetalLengthCm',
    y = 'PetalWidthCm',
    data = df,
    hue = 'Species',
    alpha = 0.5
)
plt.xlabel('Petal Width')
plt.ylabel('Petal Length')
plt.show()
```

# Exercício 3

Faça um histograma das features do dataset (exclusive coluna Species).

```python
for columns in df.columns:
  if columns == "Species":
    continue
  else:
    plt.figure(figsize = (10,10))
    sns.histplot(df[columns])
    plt.title(columns+' Histogram')
    plt.show()
  print()
```

# Exercício 4

Faça um boxplot das features do dataset (exclusive coluna Species).

```python
for columns in df.columns:
  if columns == 'Species':
    continue
  else:
    sns.boxplot(df[columns])
    plt.title(columns+' BoxPlot')
    plt.show()
  print()
```

# Exercício 5

Faça um heatmap das features do dataset (exclusive coluna Species);

Faça um pairplot para analisar a correlação entre as variáveis do dataset.

```python
# mapa de calor
new_df = df.drop(columns = 'Species')       # criando um novo dataframe genérico 'new_df' que não tem a coluna 'Species'
corr = new_df.corr()                        # calculando a matriz de correlação entre as variáveis e salvando em uma variável 'corr'
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(
    corr,
    mask = mask,
    cmap = 'Blues',
    annot = True
)
plt.show()
```

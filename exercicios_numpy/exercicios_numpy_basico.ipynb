# Exercício 22

Escreva uma função que receba um array NumPy e mostre na tela um conjunto de  estatísticas descritivas deste objeto (valor máximo, valor mínimo, média, mediana, variância, desvio padrão).

```python
def info_array(array):
  print('-'*5,'Estatística Descritiva','-'*5,'\n')
  print(f'Valor máximo: {array.max()}')
  print(f'Valor mínimo: {array.min()}')
  print(f'Média: {np.mean(array)}')
  print(f'Mediana: {np.median(array)}')
  print(f'Variância: {np.var(array)}')
  print(f'Desvio padrão: {np.std(array)}')


teste_array = False
while teste_array == False:

  teste_linhas = False
  while teste_linhas == False:
    linhas = input('Quantas linhas terá o seu vetor/matriz? ')
    try:
      linhas_int = int(linhas)
      if linhas_int > 0:
        teste_linhas = True
      else:
        print('Digite um número inteiro maior que 0.')
    except ValueError:
      print('Digite um número inteiro maior que 0.')
  print()

  teste_colunas = False
  while teste_colunas == False:
    colunas = input('Quantas colunas terá o seu vetor/matriz? ')
    try:
      colunas_int = int(colunas)
      if colunas_int > 0:
        teste_colunas = True
      else:
        print('Digite um número inteiro maior que 0.')
    except ValueError:
      print('Digite um número inteiro maior que 0.')

  print()
  lista_numeros = []
  for i in range(0,linhas_int):
    for j in range(0,colunas_int):
      teste_numero = False
      while teste_numero == False:
        numero = input(f'Digite um número para a posição {j+1} da linha {i+1}: ')
        try:
          numero_float = float(numero)
          lista_numeros.append(numero_float)
          teste_numero = True
        except ValueError:
          print('Somente números são aceitos.')

  teste_array = True

matriz = np.array(lista_numeros).reshape(linhas_int, colunas_int)
print('\nVetor/Matriz formado:\n',matriz,'\n')
info_array(matriz)
```

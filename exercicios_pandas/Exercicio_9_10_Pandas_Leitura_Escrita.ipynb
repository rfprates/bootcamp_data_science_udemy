Baixe o dataset anexo nesta aula, que foi gerado a partir da base FAOSTAT e refere-se a dados de produção e produtividade de commodities selecionadas para alguns países, entre 2016-2018. Siga os passos abaixo:

* Leia o dataset e armazene em uma variável;

* Imprima o shape do dataset;

* Imprima informações do dataset;

* Imprima o tipo de dado de cada coluna;

* Remova as colunas: 'Domain Code', 'Domain', 'Area Code', 'Element Code', 'Item Code', 'Year Code', 'Flag', 'Flag Description';

* Converta o dtype da coluna Value para float.

```python
# importando a biblioteca Pandas

import pandas as pd
```

```python
# lendo o dataset e armazenando em uma variável 'df'

df = pd.read_csv('/content/FAOSTAT_data_12-5-2020.csv')
```

```python
df.head()
```

```python
# imprimindo o shape do dataset

df.shape
```

Ou seja, o dataset tem 66 linhas e 14 colunas.

```python
# imprimindo as infos do dataset

df.info()
```

```python
# imprimindo o tipo de dado de cada coluna

df.dtypes
```

```python
# Remova as colunas: 'Domain Code', 'Domain', 'Area Code', 'Element Code', 'Item Code', 'Year Code', 'Flag', 'Flag Description';

colunas = ['Domain Code', 'Domain', 'Area Code', 'Element Code', 'Item Code', 'Year Code', 'Flag', 'Flag Description']

df.drop(columns=colunas, inplace = True)
```

```python
df
```

```python
# convertendo o tipo de dados da coluna 'Value' para float

df['Value'] = df['Value'].astype('float')
```

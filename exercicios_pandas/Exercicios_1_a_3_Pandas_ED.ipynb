# Exercício 3

Os dados a seguir representam as vendas diárias de um produto hipotético.

c) variância;

d) moda;

e) quantidade total vendida no período;

```python
df = pd.DataFrame(dic, index = index)

df
```

```python
# a)

df.mean()
```

```python
# b)

df.std()
```

```python
# c)

df.var()
```

```python
# d)

df.mode()
```

```python
# e)

df.sum()
```

```python
# f)

df[df['Vendas Diárias'] == df.max()[0]]
```

```python
# g)

df[df['Vendas Diárias'] == df.min()[0]]
```

# Exercício 6

Ainda com base no dataset do exercício 4 responda:

Estatísticas descritivas:

* Qual a distribuição ordenada idade dos clientes (use a função value_counts())?

* Qual a média e o desvio padrão da idade por gênero?

* Qual a proporção de homens e mulheres?

* Qual a distribuição do estado civil entre homens e mulheres?

* Qual o nível de educação por gênero?

```python
# Qual a distribuição ordenada idade dos clientes (use a função value_counts())?

df['Age'].value_counts()
```

```python
# Média da idade por gênero

df.groupby(by = 'Gender')['Age'].mean()
```

```python
# Desvio padrão da idade por gênero

df.groupby(by = 'Gender')['Age'].std()
```

```python
# outra forma de fazer as operações acima

metricas = [np.mean, np.std]
group = df.groupby(by = 'Gender')['Age']
group.agg(metricas)
```

```python
# Proporção de homens e mulheres

df['Gender'].value_counts()
```

```python
# Qual a distribuição do estado civil entre homens e mulheres?

group1 = df.groupby(by = 'Gender')['MaritalStatus']
group1.value_counts()
```

```python
# Qual o nível de educação por gênero?

group2 = df.groupby(by = 'Gender')['Education']
group2.value_counts()
```

Visualizando os dados
* Gere um histograma para a coluna idade;

```python
# Histograma para a coluna idade

df['Age'].plot(kind = 'hist', bins = 30)
plt.title('Histograma da idade dos clientes')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

# Exercício 1

O banco de dados possui informações sobre a distribuição de respiradores para tratamento da Covid-19 no Brasil. Com base neste dataset, faça:

* Leia o dataset com a função pd.read_csv() e armazene em uma variável df;

* Imprima as 10 primeiras linhas;

* Imprima as colunas do dataset;

* Mostre informações sobre o objeto df;

* Mostre o tipo de dado de cada coluna;

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/distribuicao_respiradores_BR.csv')
```

Percebe-se que neste caso, o dataset está com as informações separadas por ponto e vírgula. Neste caso, precisamos incluir o argumento "sep = ' ; ' " dentro do método pd.read_csv.

```python
df = pd.read_csv('/content/distribuicao_respiradores_BR.csv', sep = ';')
```

```python
# imprimindo as primeiras 10 linhas novamente

df.head(10)
```

Agora sim temos todas as informações das colunas e linhas separadas devidamente.

```python
# imprimindo as 3 últimas linhas

df.tail(3)
```

```python
# imprimindo as colunas do dataset

df.columns
```

```python
# mostrando o tipo de dados de cada coluna

df.dtypes
```

```python
# mostrando o número de linhas do dataset

df.shape[0]
```

```python
# mostrando a quantidade de missing data por coluna

df.isnull().sum()
```

# Exercício 2

Ainda com o dataset anterior siga os passos abaixo:

* Converta a coluna DATA para datetime;

* Converta a coluna VALOR para float;

* Mostre os valores únicos da coluna FORNECEDOR;

* Mostre os valores únicos da coluna TIPO;

* Mostre a quantidade de destinatários únicos.

```python
# convertendo a coluna DATA para datetime

df['DATA'] = pd.to_datetime(df['DATA'])
```

```python
# convertendo a coluna VALOR para float

df['VALOR'] = df['VALOR'].astype('float')
```

```python
# mostrando os valores únicos da coluna FORNECEDOR

df['FORNECEDOR'].unique()
```

```python
# mostrando os valores únicos da coluna TIPO

df['TIPO'].unique()
```

```python
# mostrando a quantidade de destinatários únicos

df['DESTINATARIO'].unique().size
```

# Exercício 3

Após as operações efetuadas nos exercícios 1 e 2, salve o banco de dados no formato CSV com o nome 'distribuicao_respiradores-modificado.csv'. Use ';' como separador.

```python
# salvando o dataset em outro arquivo csv

df.to_csv('distribuicao_respiradores-modificado.csv', index = False, sep = ';')
```

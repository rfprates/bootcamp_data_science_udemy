# Exercício 17

```python
def missing_data(data_frame):
  if isinstance(data_frame, pd.core.frame.DataFrame):
    missing = ((data_frame.isna().sum())/data_frame.shape[0])*100
    return missing
  else:
    return

data = {
  'A': [10, np.nan, 35, 47, 17, 13],
  'B': ['A', np.nan, 'B', 'A', 'A', np.nan],
  'C': [139, np.nan, np.nan, 199, 111, 140]
}

df = pd.DataFrame(data)

missing_data(df)
```

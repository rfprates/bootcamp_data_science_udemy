# Exercício 14

Use o DataFrame abaixo e calcule as médias aritimética, geométrica e harmônica de cada coluna. Use a biblioteca scipy.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': np.random.randint(1, 50, 10),
                   'B': np.random.randint(1, 100, 10),
                   'C': np.random.randint(1, 250, 10),
                   'D': np.random.randint(1, 120, 10)
                  })

df
```

```python
from scipy.stats import gmean, hmean    # importando da biblioteca scipy.stats as funções gmean (calcula média geométrica) e hmean (calcula média harmônica)
```

```python
# média geométrica

geo_media = pd.DataFrame(gmean(df).reshape(4,1).round(2), index = list('ABCD'), columns = ['Média Geométrica'])
geo_media
```

```python
# média harmônica

har_media = pd.DataFrame(hmean(df).reshape(4,1).round(2), index = list('ABCD'), columns = ['Média Harmônica'])
har_media
```

```python
# média aritimética

arit_media = pd.DataFrame(df.mean(), columns = ['Média Aritimética'])
arit_media
```

```python
# concatenando todos os data frames

medias = pd.concat([arit_media, geo_media, har_media], axis = 1)
medias
```

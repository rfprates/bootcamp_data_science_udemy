```python
# removendo as colunas 'Domain Code', 'Domain', 'Area Code', 'Element Code', 'Item Code', 'Year Code', 'Flag', 'Flag Description'

colunas = ['Domain Code', 'Domain', 'Area Code', 'Element Code', 'Item Code', 'Year Code', 'Flag', 'Flag Description']

df.drop(columns = colunas, inplace = True)
```

```python
# criando uma funÃ§Ã£o para a coluna 'Area', de forma a converter o nome do paÃ­s pela sua respectiva sigla

def sigla(country):
  if country == 'Argentina':
    return 'AR'
  elif country == 'Brazil':
    return 'BR'
  else:
    return 'USA'

df['Area'].apply(sigla)
```

```python
df['Siglas'] = df['Area'].apply(sigla)
```

```python
# criando uma nova coluna 'Country/Commodity' que corresponde a concatenaÃ§Ã£o das colunas 'Area' e 'Item'

df['Country/Commodity'] = df['Area']+' '+df['Item']
```

```python
df
```

```python
# agrupando os dados em produÃ§Ã£o e produtividade (coluna 'Element') e armazenando em variÃ¡veis distintas

# a funÃ§Ã£o get_group() utilizada faz com que os dados sejam agrupados com base em um valor especÃ­fico que esteja dentro da coluna especificada pela funÃ§Ã£o groupby.
# assim, no exemplo abaixo, criamos a variÃ¡vel producao e armazenamos nela o dataframe df agrupado pela coluna 'Element', onde na 'Element' tenhamos apenas
# dados iguais a 'Production
# jÃ¡ para a variÃ¡vel produtividade, agrupamos o df pela coluna 'Element', onde temos apenas dados iguais a 'Yield'


producao = df.groupby(by = 'Element').get_group('Production')
produtividade = df.groupby(by = 'Element').get_group('Yield')
```

```python
producao
```

```python
produtividade
```

```python
# Mude o shape dos dataframes criados no exercÃ­cio anterior, com a funÃ§Ã£o pivot, de modo que os novos objetos tenham Ã­ndice 'Year', coluna 'Country/Commodity'
# e os valores correspondam a coluna 'Value'

producao = producao.pivot(index= 'Year', columns = 'Country/Commodity', values = 'Value')
produtividade = produtividade.pivot(index = 'Year', columns = 'Country/Commodity', values = 'Value')
```

```python
producao
```

```python
produtividade
```

```python
# concatenando ambos os dataframes criados

df_mod = pd.concat([producao, produtividade], axis = 0, keys = ['Production', 'Yield'])
```

```python
# fazendo um mapa de calor com base na matriz de correlação, utilizando a biblioteca Seaborn

corr = df.corr()

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize = (12,12))

sns.heatmap(
    data = corr,
    annot = True,
    mask = mask,
    square = True,
    cmap = 'Blues'
);
```

```python
# mostrando o mapa de calor completo

plt.figure(figsize = (13,13))
sns.heatmap(
    data = corr,
    annot = True,
    cmap = 'Blues'
);
```

```python
# definindo a coluna 'price' como a nossa variável target

y = df['price'].values.reshape(-1,1)
```

```python
# definindo as demais colunas do dataset como variáveis preditoras

x = df.drop('price', axis = 1)
```

```python
# importando a função tran_test_split para separar o nosso dataset em base de treino e base de teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3, random_state = 33)
```

```python
# criando o modelo de regressão linear utilizando a função LinearRegression da biblioteca sklearn.linear_model

from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(x_treino, y_treino)
```

```python
# calculando o coeficiente de determinação do modelo

modelo.score(x_teste, y_teste)
```

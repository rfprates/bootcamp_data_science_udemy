```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Exercício 15

* Carregue o dataset Iris anexo nesta aula.

* Remova a coluna Id;

* Faça um gráfico de barras para mostrar a frequência das três espécies de Iris do dataset;

* Faça um gráfico de barras horizontal para obter o resultado anterior.

```python
df = pd.read_csv('/content/Iris.csv')
```

```python
df
```

```python
df.drop(columns = 'Id', inplace = True)
```

```python
df
```

```python
plt.bar(x = df['Species'].unique(), height = df['Species'].value_counts());
plt.title('Frequência de cada uma das espécies')
plt.xlabel('Espécies')
plt.ylabel('Frequência')
plt.show()
```

```python
# gráfico de barras horizontal

plt.barh(y = df['Species'].unique(), width = df['Species'].value_counts())
plt.title('Frequência de cada uma das espécies em barras horizontais')
plt.xlabel('Frequência')
plt.ylabel('Espécies');
```

# Exercício 16

Faça um gráfico de pizza para mostrar a frequência das três espécies de Iris do dataset;

```python
plt.pie(x = df['Species'].value_counts(), autopct = '%1.2f%%')
plt.title('Frequência das espécies de Iris')
plt.legend(labels = df['Species'].unique(), loc = 'upper left')
plt.show()
```

# Exercício 17

Faça um scatter plot para mostrar a relação entre 'SepalLengthCm' e 'SepalWidthCm';

```python
plt.scatter(df['SepalLengthCm'],df['SepalWidthCm'])
plt.title('Relationship between Sepal Lenght and Sepal Width')
plt.xlabel('Sepal Lenght (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(ls = '--', alpha = 0.5)
plt.show()
```

# Exercício 18

Faça um scatter plot para mostrar a relação entre 'PetalLengthCm' e 'PetalWidthCm'.

```python
plt.scatter(df['PetalLengthCm'],df['PetalWidthCm'])
plt.title('Relationship between Petal Length and Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.grid(ls = '--', alpha = 0.5)
plt.show()
```

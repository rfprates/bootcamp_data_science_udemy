# Exercício 1

Use os arrays a seguir para construir um gráfico de linhas. Use a biblioteca Matplotlib e configure o título com plt.title(), e a descrição dos eixos com plt.xlabel() e plt.ylabel().

# Exercício 4

Use o DataFrame abaixo para construir um gráfico de linhas, para os valores da coluna 'Brand 1' em função do tempo.

```python
n = 100
columns = ['Brand'+str(number) for number in range(1,5)]
date = pd.date_range(start = '2021-1-1', periods = n)
data = np.random.randint(low = 1, high = [100,330,45,77], size = (n,4))
```

```python
df
```

```python
plt.figure(figsize = (14,6))
plt.plot(df['Brand1'])
plt.title('Brand1')
plt.xlabel('Date')
plt.ylabel('Values');
```

# Exercício 5

Construa um gráfico de barras com o DataFrame do exercício anterior. Use a função plt.bar()

```python
# somando os dados por marca

data_soma = df.sum()
data_soma
```

```python
# pegando os índices para atribuir ao eixo x do gráfico

x = data_soma.index
x
```

```python
# plotando o gráfico

plt.bar(x = x, height = data_soma)
plt.title('Title')
plt.xlabel('Brands')
plt.ylabel('Values');
```

```python
# para ordenar o gráfico do maior para o menor

data_soma = data_soma.sort_values(ascending = False)
x = data_soma.index
plt.bar(x = x, height = data_soma);
```

# Exercício 7

Use o código abaixo para construir um scatter plot. Use a função plt.scatter() e configure o marcador para '^' (marker = '^') e a opacidade para 0.3 (alpha = 0.3). Use plt.grid() configurar as grid lines.

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples = 100, n_features = 1, noise = 20)
```

```python
plt.scatter(X,y, marker = '^', alpha = 0.3)
plt.title('Relationship between X and Y')
plt.ylabel('Target')
plt.xlabel('Feature')
plt.grid(ls = '--', alpha = 0.2)
plt.savefig('fig7.jpeg')    # comando para salvar o gráfico em um arquivo de imagem
plt.show()
```

# Exercício 9

Use os dados a seguir para construir o gráfico abaixo. Altere o tamanho da fonte para 15 (fontsize) nos eixos e 20 no título. Use a função plt.legend() para inserir uma legenda no gráfico, use o argumento loc para configurar a posição. Possibilidades para a configuração do argumento loc: best, upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center.

```python
x = np.arange(1,11)
y1 = x**2
y2 = x**3
y3 = x**4
```

```python
plt.plot(x,y1,'o', x,y2,'^', x,y3,'bs')
plt.title('Values', fontsize = 20)
plt.xlabel('X', fontsize = 15)
plt.legend(labels = ['y1','y2','y3'], loc = 'upper left')
plt.grid(ls = '--', alpha = 0.3);
```

# Exercício 11

Use os dados abaixo para construir um gráfico de pizza. Use a função plt.pie() e salve a figura obtida.

```python
sales = [1500.2,1378,671,431,700]
labels = ['Brand '+ str(number) for number in range(1,len(sales)+1)]
```

```python
plt.pie(x = sales, labels = labels);
plt.savefig('Grafico_pizza_11.jpeg')
```

# Exercício 12

Volte no gráfico anterior e adicione legendas com a função plt.legend() (loc = 'upper left'), mude o argumento shadow para True e mostre o percentual de vendas de cada marca.

```python
explode = (0.1,0.1,0,0,0)

plt.pie(x = sales, labels = labels, shadow = True, autopct = '%1.2f%%', explode = explode)
plt.legend(labels = labels, loc = 'upper left');
```

# Exercício 13

Construa o DataFrame a seguir. Posteriormente elabore o gráfico abaixo. Mude o limite do eixo y para 80-100, com a função plt.ylim(), e o tamanho da figura para (15,5) (use plt.figure(figsize = (15,5)).

```python
np.random.seed(7)
data = np.random.randint(low = 100, high = [250,300,500,1000], size = (12,4))
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

df = pd.DataFrame(data, index = months, columns = list('ABCD'))
```

```python
df
```

```python
plt.figure(figsize = (15,5))
plt.plot(df)
plt.ylim((80,1000))
plt.legend(labels = df.columns, loc = 'upper left')
plt.grid(ls = '--', alpha = 0.3);
```

# Exercício 14

Com o DataFrame anterior construa: a) gráfico de barras; b) gráfico de barras horizontal; c) gráfico de pizza.

```python
# a)

plt.bar(x = df.columns, height = df.sum())
plt.title('Bar plot')
plt.show()
```

```python
# b)

plt.barh(y = df.columns, width = df.sum())
plt.title('Horizontal bar plot');
```

```python
# c)
explode = (0, 0.1, 0.1, 0.1)
plt.pie(x = df.sum(), labels = df.columns, autopct = '%1.2f%%', explode = explode, shadow = True)
plt.legend(labels = df.columns, loc = 'upper left')
plt.show()
```

```python
import pandas as pd
import yfinance as yf
```

```python
df = yf.Ticker('PETR4.SA').history(period = 'max')
```

```python
df
```

```python
df.info()
```

```python
df.head()
```

```python
# mostrando os dados das colunas 'Open' e 'Close' para os dias entre novembro e dezembro de 2020

df.loc['2020-11-01':'2020-12-31', ['Open', 'Close']]
```

```python
# mostrando os dias em que o preÃ§o de abertura foi superior R$ 32.95

df[df['Open']>32.95]
```

```python
# mostrando os dias em que o preÃ§o de abertura foi maior ou igual a 30 e o preÃ§o mÃ¡ximo foi superior a 33.5

df[(df['Open']>=30) & (df['High']>33.5)]
```

```python
# mostrando os dias em que o preÃ§o mÃ¡ximo foi superior a 32.8 e inferior a 33

df[(df['Close']>32.8) & (df['Close']<33)]
```

```python
# mostrando os dias em que o preÃ§o mÃ¡ximo foi inferior a 7 ou o preÃ§o mÃ­nimo foi inferior a 9

df[(df['High']<7) | (df['Low']<9)]
```

```python
# convertendo o data frame para csv

df.to_csv('financas.csv', index = True)
```
